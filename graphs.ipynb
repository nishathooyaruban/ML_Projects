```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load CSV
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\HSQS.csv")
```

```python
df.head()
```

```python
# checking the data info
print(' df data info:\n')
df.info()
```

```python
sns.scatterplot(x='Readmission_Rate (%)', y='Patient_Satisfaction (%)', data=df)
plt.title('Patient Satisfaction vs. Readmission Rate')
plt.xlabel('Readmission Rate (%)')
plt.ylabel('Patient Satisfaction (%)')
plt.show()
```

```python
sns.scatterplot(x='Operating_Costs ($)', y='Quality_Score', data=df)
plt.title('Operating_Costs vs. Quality_Score')
plt.xlabel('Operating_Costs ($)')
plt.ylabel('Quality_Score')
plt.show()
```

```python
# Create categories for Readmission Rate
df['Readmission_Category'] = pd.cut(df['Readmission_Rate (%)'], bins=[0, 10, 20, 30], labels=['Low', 'Medium', 'High'])

# Create count plot
sns.countplot(x='Readmission_Category', data=df, palette='viridis')
plt.title('Count of Hospitals by Readmission Rate Category')
plt.xlabel('Readmission Rate Category')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
# Create categories for Quality Score
df['Quality_Category'] = pd.cut(df['Quality_Score'], bins=[0, 70, 80, 90], labels=['Low', 'Medium', 'High'])

# Create count plot
sns.countplot(x='Quality_Score', data=df, palette='magma')
plt.title('Count of Hospitals by Quality Score Category')
plt.xlabel('Quality Score Category')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
# Create categories for Staff-to-Patient Ratio
df['Staff_Ratio_Category'] = pd.cut(df['Staff_to_Patient_Ratio'], bins=[0, 3.5, 4.5, 5], labels=['Low', 'Medium', 'High'])

# Create count plot
sns.countplot(x='Staff_Ratio_Category', data=df, palette='plasma')
plt.title('Count of Hospitals by Staff-to-Patient Ratio Category')
plt.xlabel('Staff-to-Patient Ratio Category')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
# Create categories for Patient Satisfaction
df['Satisfaction_Category'] = pd.cut(df['Patient_Satisfaction (%)'], bins=[0, 70, 80, 90, 100], labels=['Low', 'Medium', 'High', 'Very High'])

# Create count plot
sns.countplot(x='Satisfaction_Category', data=df, palette='coolwarm')
plt.title('Count of Hospitals by Patient Satisfaction Category')
plt.xlabel('Patient Satisfaction Category')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
# Create categories for Operating Costs
df['Cost_Category'] = pd.cut(df['Operating_Costs ($)'], bins=[1000000, 1500000, 2000000], labels=['Low', 'High'])

# Create count plot
sns.countplot(x='Cost_Category', data=df, palette='Set2')
plt.title('Count of Hospitals by Operating Cost Category')
plt.xlabel('Operating Cost Category')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
# Create categories for Average Length of Stay
df['Stay_Category'] = pd.cut(df['Avg_Length_of_Stay '], bins=[0, 4, 6], labels=['Short', 'Long'])

# Create count plot
sns.countplot(x='Stay_Category', data=df, palette='rocket')
plt.title('Count of Hospitals by Average Length of Stay Category')
plt.xlabel('Average Length of Stay Category')
plt.ylabel('Number of Hospitals')
plt.show()
```

```python
sns.countplot(x='Readmission_Category', hue='Quality_Category', data=df, palette='viridis')
plt.title('Count of Hospitals by Readmission Rate and Quality Score Categories')
plt.xlabel('Readmission Rate Category')
plt.ylabel('Number of Hospitals')
plt.legend(title='Quality Score Category')
plt.show()
```

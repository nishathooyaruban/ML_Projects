```python
from sklearn.model_selection import train_test_split
x = df.drop(['MEDV'], axis=1)
y = df['MEDV']
train_features, test_features, train_labels, test_labels = train_test_split(x, y, test_size=0.2, random_state=54545)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler() 
train_features_scaled = std_scaler .fit_transform (train_features)
test_features_scaled  =std_scaler.fit_transform (test_features)
```

```python
from sklearn.linear_model import LinearRegression
model_LR = LinearRegression()
model_LR.fit (train_features_scaled,train_labels)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import math

def performance_reg(model, test_features, test_labels):
    pred = model.predict(test_features)
    print(f"R2 score : {r2_score(test_labels, pred)}")
    print(f"MSE score : {mean_squared_error(test_labels, pred)}")
    print(f"RMSE score : {math.sqrt(mean_squared_error(test_labels, pred))}")
    print("------------------------------------------------------")
```

```python
performance_reg(model_LR,test_features_scaled,test_labels)
```
